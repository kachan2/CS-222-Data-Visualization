```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/jash_fa_17_cleaned.csv', sep=',')
df
```

```python
ax1 = plt.scatter(df["zip"], df["Total"],)
df092 = df.loc[(df['zip'] > 50000) &  (df['zip'] < 70000)]
ax2 = plt.scatter(df092["zip"], df092["Total"],)
```

```python
states = df['state_name']
df2 = df.groupby('state_name')['Total'].sum()
df2= pd.DataFrame(df2)



states = states.T.drop_duplicates().T
states = pd.DataFrame(states)
states = states.sort_values('state_name')
plt.pie(df2['Total'], labels = states['state_name'])
```
